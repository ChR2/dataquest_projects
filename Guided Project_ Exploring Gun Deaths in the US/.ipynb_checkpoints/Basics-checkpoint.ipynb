```python
import csv
import datetime

#guns = open("guns.csv", "r")
#csv_data = csv.reader(guns)
#gun_data = list(csv_data)

with open("guns.csv") as gun_data:
    data = list(csv.reader(gun_data))
```

```python
header = data[0]
data = data[1:]
print(data[:5])
```

```python
years = [row[1] for row in data]
print(years)
```

```python
year_counts = {}
for year in years:
    if year in year_counts:
        year_counts[year] += 1
    else:
        year_counts[year] = 1
print(year_counts)
```

```python
dates =[datetime.datetime(year = int(row[1]), month = int(row[2]), day =1) for row in data] 
```

```python
sex_counts = {}
for row in data:
    if row[5] in sex_counts:
        sex_counts[row[5]] += 1
    else:
        sex_counts[row[5]] = 1
print(sex_counts)
```

```python
race_counts = {}
for row in data:
    if row[7] in race_counts:
        race_counts[row[7]] += 1
    else:
        race_counts[row[7]] = 1
print(race_counts)
```
